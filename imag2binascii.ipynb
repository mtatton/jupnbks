```python
#
# https://www.iconexperience.com/
#

from matplotlib.pyplot import imshow
import numpy as np
from PIL import Image

%matplotlib inline

def image2binsci(path,fname,ext):
  pil_im = Image.open(path+'\\'+fname+'.'+ext, 'r')
  imshow(np.asarray(pil_im))
  out_arr=np.asarray(pil_im)
  out_sum=np.sum(out_arr,axis=2)
  out_ans=np.where(out_sum > 0,1,0)
  with open(path+'\\'+fname+'.'+'txt', 'wb') as f:
    np.savetxt(f, out_ans, delimiter='', newline='\n', header='', footer='', comments='# ',fmt='%s')
    
path='.'
fname='debian'
ext='png'
image2binsci(path,fname,ext)
```
